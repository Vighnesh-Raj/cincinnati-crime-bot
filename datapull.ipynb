```python
import pandas as pd

from sodapy import Socrata
import time
from datetime import datetime
import os

# Set dataset and output config
DOMAIN = "data.cincinnati-oh.gov"
DATASET_ID = "gexm-h6bt"
LIMIT = 50000  # Max allowed for SODA 2.0 API
OUTPUT_DIR = "data"
DATESTAMP = datetime.now().strftime("%Y%m%d")
FILENAME = f"calls_for_service_{DATESTAMP}.csv"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Create Socrata client (unauthenticated for public data)
client = Socrata(DOMAIN, None)

def fetch_all_data():
    offset = 0
    all_results = []

    while True:
        print(f"Fetching records {offset} to {offset + LIMIT}...")
        results = client.get(DATASET_ID, limit=LIMIT, offset=offset)
        
        if not results:
            print("No more records to fetch.")
            break

        all_results.extend(results)
        offset += LIMIT
        time.sleep(0.5)  # prevent overloading API

    print(f" Total records fetched: {len(all_results)}")
    df = pd.DataFrame.from_records(all_results)

    filepath = os.path.join(OUTPUT_DIR, FILENAME)
    df.to_csv(filepath, index=False)
    print(f" Data saved to: {filepath}")

    return df

if __name__ == "__main__":
    fetch_all_data()
```
